```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

class LeetCodeScraper:
    def __init__(self):
        self.columns = ['Question ID', 'Question Title', 'Question Slug', 'Question Text',
                        'Topic Tagged text', 'Difficulty Level', 'Success Rate',
                        'total submission', 'total accepted', 'Likes', 'Dislikes',
                        'Hints', 'Similar Questions ID', 'Similar Questions Text']
        self.data = {col: [] for col in self.columns}
        self.base_url = "https://leetcode.com/problems/"

    def scrape_question(self, question_slug):
        url = f"{self.base_url}{question_slug}/"

        response = requests.get(url)
        if response.status_code != 200:
            print(f"Failed to fetch data for {question_slug}")
            return

        soup = BeautifulSoup(response.text, "html.parser")

        question_id = question_slug
        question_title = str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        question_slug = question_slug
        question_text = str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        topic_tagged_text = str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        difficulty_level =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        success_rate =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        total_submission = str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        total_accepted =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        likes =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        dislikes =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        hints =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        similar_questions_id =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))
        similar_questions_text =  str(soup.find("a", attrs={"class": "mr-2 text-label-1 dark:text-dark-label-1 hover:text-label-1 dark:hover:text-dark-label-1 text-lg font-medium"}))

        self.data['Question ID'].append(question_id)
        self.data['Question Title'].append(question_title)
        self.data['Question Slug'].append(question_slug)
        self.data['Question Text'].append(question_text)
        self.data['Topic Tagged text'].append(topic_tagged_text)
        self.data['Difficulty Level'].append(difficulty_level)
        self.data['Success Rate'].append(success_rate)
        self.data['total submission'].append(total_submission)
        self.data['total accepted'].append(total_accepted)
        self.data['Likes'].append(likes)
        self.data['Dislikes'].append(dislikes)
        self.data['Hints'].append(hints)
        self.data['Similar Questions ID'].append(similar_questions_id)
        self.data['Similar Questions Text'].append(similar_questions_text)

    def scrape_questions_list(self, question_slugs):
        for slug in question_slugs:
            self.scrape_question(slug)

    def create_dataframe(self):
        df = pd.read_csv('/content/drive/MyDrive/sem-9/IR/leetcode_questions.csv')
        return df
```
